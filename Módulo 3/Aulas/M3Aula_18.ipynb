```python
from tensorflow import keras
from keras.datasets import cifar10

print("Loading CIFAR10 images ...")
(Xtrain, Ytrain), (Xtest, Ytest) = cifar10.load_data()

Ytrain = keras.utils.to_categorical(Ytrain)
Ytest = keras.utils.to_categorical(Ytest)

print('Original X training set shape: ', Xtrain.shape)
print('Original Y training set shape: ', Ytrain.shape)
print('Original X testing set shape: ', Xtest.shape)
print('Original Y testing set shape: ', Ytest.shape)
```

```python
model = keras.Sequential(
    [keras.layers.Conv2D(8, (3, 3), padding='same', name='conv1', input_shape=((32,32,3))),
    keras.layers.Conv2D(8, (3, 3), padding='same', name='conv2'),
    keras.layers.MaxPooling2D(pool_size=(2, 2), name='pool1'),
    keras.layers.Flatten(name='code'),
    keras.layers.Dense(10, activation='softmax', name='pred'),
    ]
)
model.summary()

model.compile(loss="categorical_crossentropy", metrics='accuracy')
```

```python
history1 = model.fit(Xtrain, Ytrain, batch_size=32, epochs=30)

print("\n")
scoreCNN = model.evaluate(Xtest, Ytest)
print(model.metrics_names)
print("Acurácia: ", scoreCNN[1])
```

# Qual o valor da loss nas épocas 10, 20, 30, 40 e 50? O que podemos interpretar destes valores obtidos?

```python
print("Loading CIFAR10 images ...")
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

y_train = keras.utils.to_categorical(y_train)
y_test = keras.utils.to_categorical(y_test)

print('Original X training set shape: ', x_train.shape)
print('Original Y training set shape: ', y_train.shape)
print('Original X testing set shape: ', x_test.shape)
print('Original Y testing set shape: ', y_test.shape)
```

```python
model3 = keras.Sequential(
    [keras.layers.Conv2D(16, (5, 5), padding='same', name='conv1', input_shape=((28, 28, 1))),
     keras.layers.Conv2D(16, (5, 5), padding='same', name='conv2'),
     keras.layers.Conv2D(16, (5, 5), padding='same', name='conv3'),     
     keras.layers.AveragePooling2D(pool_size=(2, 2), name='pool1'),     
     keras.layers.Flatten(name='code'),
     keras.layers.Dense(10, activation='softmax', name='pred'),
    ]
)
model3.summary()

model3.compile(loss="categorical_crossentropy", metrics='accuracy')

history2 = model3.fit(x_train, y_train, batch_size=32, epochs=50)

print("\n")
scoreCNN = model3.evaluate(x_test, y_test)
print(model3.metrics_names)
print("Acurácia: ", scoreCNN[1])
```

Loss época 10 = 0.3688  
Loss época 20 = 0.3645   
Loss época 30 = 0.3593  
Loss época 40 = 0.3557  
Loss época 50 = 0.3661  

Não há grandes mudanças no valor de loss, isso significa que o modelo converge (torna-se praticamente constante) em épocas anteriores a 10.
