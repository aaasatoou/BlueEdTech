```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
X, y = digits.data, digits.target

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size = 0.2,random_state=10)
modelo_DT = DecisionTreeClassifier()

kf = KFold(n_splits=5)
score = cross_val_score(modelo_DT, X, y, cv=kf, scoring='accuracy')
print(f"{score.mean()} ± {score.std()}")

modelo_DT = modelo_DT.fit(X_train, y_train)
```

```python
plt.figure(figsize=(14,14))
plot_tree(modelo_DT, filled=True)
plt.show()
```

```python
modelo_DT_imp = DecisionTreeClassifier(min_impurity_decrease=0.02)

score_imp = cross_val_score(modelo_DT_imp, X, y, cv=kf, scoring='accuracy')
print(f"{score_imp.mean()} ± {score_imp.std()}")

modelo_DT_imp = modelo_DT_imp.fit(X_train, y_train)

plt.figure(figsize=(14,14))
plot_tree(modelo_DT_imp, filled=True)
plt.show()
```

```python
modelo_DT_ent = DecisionTreeClassifier(criterion='entropy', max_depth=5)

score_ent = cross_val_score(modelo_DT_ent, X, y, cv=kf, scoring='accuracy')
print(f"{score_ent.mean()} ± {score_ent.std()}")

modelo_DT_ent = modelo_DT_ent.fit(X_train, y_train)

plt.figure(figsize=(40,40))
plot_tree(modelo_DT_ent, filled=True)
plt.show()
```
