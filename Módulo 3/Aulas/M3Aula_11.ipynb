1. Quantos substantivos e verbos foram identificados?

```python
# verbo == VERB
# substantivo == NOUN

verbo = 0
substantivo = 0

for token in texto_ex1:

  if token.pos_ == 'VERB':
    verbo += 1

  elif token.pos_ == 'NOUN':
    substantivo += 1

print(f'Verbos: {verbo}')
print(f'Substantivos: {substantivo}')
```

2. Separe todos os verbos em uma lista e aplique a Lemmatization sobre eles obtendo outra lista. Nesta nova lista remova as duplicatas e conte quantos verbos únicos há.

```python
verbo_list =[]
lemma_list = []

for token in texto_ex1:
  if token.pos_ == 'VERB':
    verbo_list.append(token.text.lower())
    lemma_list.append(token.lemma_.lower())

print(verbo_list)
print(lemma_list)
```

```python
# Remoção de duplicatas

lemma_list = list(dict.fromkeys(lemma_list))
print(lemma_list)
```

```python
print(f'{len(lemma_list)} verbos únicos')
```

2. Adicione as entidades que faltaram e verifique se agora eles serão identificados.

A entidades foram identificadas corretamente, porém, Tesla foi interpretado como uma pessoa (Nikola Tesla) e não a empresa Tesla. Assim como Sephora e Louis Vuitton

```python
ents = list(texto_ex2.ents)
old_ent = ents[0]
new_ent = Span(texto_ex2, old_ent.start, old_ent.end, label="ORG")
ents[0] = new_ent

old_ent = ents[3]
new_ent = Span(texto_ex2, old_ent.start, old_ent.end, label="ORG")
ents[3] = new_ent

old_ent = ents[5]
new_ent = Span(texto_ex2, old_ent.start, old_ent.end, label="ORG")
ents[5] = new_ent

old_ent = ents[6]
new_ent = Span(texto_ex2, old_ent.start, old_ent.end, label="ORG")
ents[6] = new_ent


texto_ex2.ents = ents
```
