```python
import numpy as np
import pandas as pd

from tensorflow import keras
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_breast_cancer
```

```python
print('MODELO DE 3 CAMADAS:')

breast_cancer = load_breast_cancer()
XMLP = breast_cancer.data
YMLP = breast_cancer.target

xTrainMLP, xTestMLP, yTrainMLP, yTestMLP = train_test_split(XMLP, YMLP, test_size=0.2)

yTrainMLP = keras.utils.to_categorical(yTrainMLP)
yTestMLP = keras.utils.to_categorical(yTestMLP)

print(yTrainMLP.shape)
print(yTestMLP.shape)


model = keras.Sequential(
    [
    keras.layers.Dense(xTrainMLP.shape[-1], activation="relu", input_shape=(xTrainMLP.shape[-1],)),
     
    keras.layers.Dense(90, activation="sigmoid"),
    keras.layers.Dense(90, activation="sigmoid"),
    keras.layers.Dense(90, activation="sigmoid"),
          
    keras.layers.Dense(2, activation="softmax"),
    ]
)

model.summary()

model.compile(loss="categorical_crossentropy", metrics='accuracy')
history1 = model.fit(xTrainMLP, yTrainMLP, batch_size=10, epochs=100, verbose=False)

print("\n")
scoreMLP = model.evaluate(xTestMLP, yTestMLP)
print(model.metrics_names)
print("Acurcia: ", scoreMLP[1])
```

```python
# visualizando a convergncia da rede durante o treinamento
import matplotlib.pyplot as plt
plt.plot(history1.history["loss"])
plt.title("model loss")
plt.ylabel("loss")
plt.xlabel("epoch")
plt.show()
```
